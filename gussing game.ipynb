```python
from random import randint
level = 1
chances = 10
total =0
while level <= 10 :
    tn = randint(1, 10)
    score = 100
    step = 100 / chances
    print("you have a number of chances = ", chances)
    cnt = 0
    while cnt < chances :
        gn = int(input ("guess a number "))
        if gn==tn:
            print("you win ,,,right guess")
            print("up the second level  ")
            print ("score = ",score )
            total += score
            print ("your total score is ", total )
            level += 1
            chances -= 1
            break

        else :
            print("wrong guess try again ")
            score = score - step
        cnt += 1
    else :
        print(" complete the same level GaMe OVER ")
```
